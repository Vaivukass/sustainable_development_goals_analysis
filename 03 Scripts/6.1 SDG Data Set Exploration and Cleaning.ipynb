## 4. Data Cleaning

### 4.1 Renaming Columns

```python
# Renaming columns to be able to distinguish among goals with keywords

df.rename(columns = {'sdg_index_score' : 'sdg_index'}, inplace = True)
df.rename(columns = {'goal_1_score' : 'no_poverty'}, inplace = True)
df.rename(columns = {'goal_2_score' : 'zero_hunger'}, inplace = True)
df.rename(columns = {'goal_3_score' : 'good_health'}, inplace = True)
df.rename(columns = {'goal_4_score' : 'quality_education'}, inplace = True)
df.rename(columns = {'goal_5_score' : 'gender_equality'}, inplace = True)
df.rename(columns = {'goal_6_score' : 'clean_water'}, inplace = True)
df.rename(columns = {'goal_7_score' : 'affordable_energy'}, inplace = True)
df.rename(columns = {'goal_8_score' : 'economic_growth'}, inplace = True)
df.rename(columns = {'goal_9_score' : 'infrastructure_innovation'}, inplace = True)
df.rename(columns = {'goal_10_score' : 'reduced_inequalities'}, inplace = True)
df.rename(columns = {'goal_11_score' : 'sustainable_cities'}, inplace = True)
df.rename(columns = {'goal_12_score' : 'responsible_consumption'}, inplace = True)
df.rename(columns = {'goal_13_score' : 'climate_action'}, inplace = True)
df.rename(columns = {'goal_14_score' : 'underwater_life'}, inplace = True)
df.rename(columns = {'goal_15_score' : 'land_life'}, inplace = True)
df.rename(columns = {'goal_16_score' : 'peace_justice'}, inplace = True)
df.rename(columns = {'goal_17_score' : 'partnerships'}, inplace = True)
```

### 4.2 Dropping / Adding Columns

```python
# Dropping 'country_code' column

df = df.drop('country_code', axis=1)
```

### 4.3 Changing Data Types

```python
# Checking mixed data types in df

for col in df.columns.tolist():
      weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
      if len (df[weird]) > 0:
        print (col)
```

#### No mixed-data types
